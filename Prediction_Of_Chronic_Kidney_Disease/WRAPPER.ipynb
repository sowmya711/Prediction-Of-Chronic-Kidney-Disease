```python
for i in x.columns:
  x[i].fillna(x[i].mean(),inplace=True)
```

```python
corr_mat = data.corr()
from scipy.stats import pointbiserialr
from math import sqrt

def getMerit(subset, label):
    ks = len(subset)

    # average feature-class correlation
    rcf_all = []
    for feature in subset:
        coeff = pointbiserialr( x[label], x[feature] )
        rcf_all.append( abs( coeff.correlation ) )
    cfr = np.mean( rcf_all )

    # average feature-feature correlation
    corr = x[subset].corr()
    corr.values[np.tril_indices_from(corr.values)] = np.nan
    corr = abs(corr)
    rff = corr.unstack().mean()

    return (ks * cfr) / sqrt(ks + ks * (ks-1) * rff)
#ks-no of features
#rcf_all- list of feature-class correlations
#rcf- average feature-class correlation between feature and label
#rff- average corr between each pair of features in the subset
# (ks * cfr) / sqrt(ks + ks * (ks-1) * rff)-merit score
#how well the subset of features can predict the label
```

```python
corr_mat.values[np.tril_indices_from(corr_mat.values)] = np.nan
corr_mat
```

```python
best_value = -1
best_feature = ''
for feature in x.columns.to_list():
    coeff = pointbiserialr(data["classification"],x[feature])
    abs_coeff = abs( coeff.correlation )
    if abs_coeff > best_value:
        best_value = abs_coeff
        best_feature = feature

print("Feature %s with merit %.4f"%(best_feature, best_value))
# highest absolute correlation with a target variable "classification"
#pointbiserialr- computes measure of the association between a binary variable (the label) and a continuous variable (the feature)
```

```python
import statsmodels.api as sm
def forward_selection(data, target, significance_level=0.05):
    initial_features = data.columns.tolist()
    best_features = []
    while (len(initial_features)>0):
        remaining_features = list(set(initial_features)-set(best_features))
        new_pval = pd.Series(index=remaining_features)
        for new_column in remaining_features:
            model = sm.OLS(target, sm.add_constant(data[best_features+[new_column]])).fit()
            new_pval[new_column] = model.pvalues[new_column]
        min_p_value = new_pval.min()
        if(min_p_value<significance_level):
            best_features.append(new_pval.idxmin())
        else:
            break
    return best_features
```

```python
features = forward_selection(x,y)[:6]
```
